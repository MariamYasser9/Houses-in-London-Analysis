# Visualization

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Price (£)'], bins=30, kde=True)
plt.title('Distribution of Property Prices (£)')
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numeric_cols], orient="h", palette="Set1")
plt.title("Boxplot for All Numeric Columns (Detecting Outliers)")
plt.xlabel("Value")
plt.ylabel("Features")
plt.show()
```

```python
corr = df[['total_rooms', 'Square Meters', 'Building Age', 'Floors', 'Price (£)', "Garden", "Garage"]].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Numeric Features')
```

```python
neigh_order = (
    df.groupby("Neighborhood")["Price (£)"]
    .mean()
    .sort_values(ascending=False)
    .index
)
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x="Square Meters",
    y="Price (£)",
    hue="Neighborhood",
    hue_order=neigh_order,
    palette="tab10",
    alpha=0.7
)

plt.title("Relationship Between Property Area and Price by Neighborhood", fontsize=13, weight="bold")
plt.xlabel("Property Area (Square Meters)")
plt.ylabel("Price (£)")
plt.legend(title="Neighborhood", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
sns.swarmplot(
    x='total_rooms',
    y='Price (£)',
    data=df,
    size=4
)

plt.title("Relationship Between Total Rooms and Price (£)", fontsize=14, weight="bold", pad=15)
plt.xlabel("Total Rooms")
plt.ylabel("Price (£)")
plt.show()
```

```python
sns.scatterplot(x='Square Meters', y='Price (£)', data=df)
plt.title('Price vs. Square Meters')
```

```python
# نحدد العوامل اللي هنقارنها مع السعر
features = ['total_rooms', 'Neighborhood', 'Property Type',
             'Garage', 'Garden', 'Balcony', 'View']

# إعداد الشكل العام للرسومات
plt.figure(figsize=(16, 25))

for i, feature in enumerate(features, 1):
    plt.subplot(4, 2, i)
    sns.violinplot(x=feature, y='Price (£)', data=df, palette='viridis')
    plt.title(f'Price (£) vs {feature}', fontsize=12)
    plt.xticks(rotation=45)
    plt.xlabel(feature)
    plt.ylabel('Price (£)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# تحديد الأعمدة المطلوبة
columns = [
    "Neighborhood", "Square Meters", "Building Age", "Garden", "Garage", "Floors",
    "Property Type", "Heating Type", "Balcony", "Interior Style", "View",
    "Materials", "Building Status", "total_rooms"
]
target = "Price (£)"

# إعداد الشكل العام للرسومات
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (7, 4)

# نمر على كل عمود ونعمل الرسم المناسب
for col in columns:
    plt.figure()

    # لو العمود رقمي → Scatterplot
    if df[col].dtype in ['int64', 'float64']:
        sns.scatterplot(data=df, x=col, y=target, color='royalblue', alpha=0.6)
        plt.title(f"Relationship between {col} and Price (£)")

    # لو العمود تصنيفي → Boxplot
    else:
        sns.boxplot(data=df, x=col, y=target, palette="pastel")
        plt.title(f"Price (£) Distribution by {col}")
        plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
```
